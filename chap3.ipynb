```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.plot([1,2,3,2,3,2,2,1])
plt.plot([4,3,2,1], [1,2,3,4])
```

```python
# some simple data
x = [1,2,3,4]
y = [5,4,3,2]

# create a new figure
plt.figure(figsize=(12,6))

# divide subplots into 2x3 grid
plt.subplot(231)
plt.plot(x, y)

# select #2
plt.subplot(232)
plt.bar(x, y)

# horizontal bar charts
plt.subplot(233)
plt.barh(x, y)

# stacked bar charts
plt.subplot(234)
plt.bar(x, y)
y1 = [7,8,5,3]
plt.bar(x, y1, bottom=y, color='r')

# bot plot
plt.subplot(235)
plt.boxplot(x)

# scatter plot
plt.subplot(236)
plt.scatter(x, y)

plt.show()
```
